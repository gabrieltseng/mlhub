We can check all the collections

````python
collections = mlhub.get_collections()
print([col["id"] for col in collections])
````

We are interested in the agricultural collections. As mentioned above, `ref_african_crops_kenya_02_*` is set up differently from the other collections.

You can download it using `mlhub`, but these exploration functions won't work as well.

````python
crop_collections = [
    "ref_african_crops_uganda_01", "ref_african_crops_tanzania_01", "ref_african_crops_kenya_01"
]
````

`plot_range` can be used to see how the source images are available over time for each feature

````python
for collection_id in crop_collections:
    print(collection_id)
    assets = mlhub.get_all_assets(collection_id)
    
    assets_df = mlhub.features_to_df(assets)
    
    dates_df = mlhub.check_dates_across_features(assets_df)
    
    mlhub.plot_range(dates_df)
````
