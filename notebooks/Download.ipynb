# Downloading datasets

```python
import mlhub
from pathlib import Path
```

To download source imagery, an API access token and AWS access keys are necessary

```python
mlhub.set_token(
    "token123"
)

mlhub.set_aws_credentials(
    aws_access_key="access123",
    aws_secret_key="secret123",
)
```

You also need to define a download location.

```python
download_location = Path("../data")
```

`download_collection` has so far been run on a few of the datasets, but mostly copies code from the [tutorial notebooks](https://github.com/radiantearth/mlhub-tutorials).

A helpful feature of this function is that it checkpoints - in this case, you can see that for each feature, `Documentation.pdf` and the labels geojson already exist, so its skipped instead of being downloaded again.

```python
mlhub.download_collection(
    collection_id="ref_african_crops_tanzania_01",
    download_path=download_location
)
```
